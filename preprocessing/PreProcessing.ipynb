## Imputate

```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('./train.csv')
test_df = pd.read_csv('./test.csv')
```

```python
train = train_df.iloc[:, 6:]
train
```

### singleimputation

```python
from sklearn.impute import SimpleImputer
train_mean = SimpleImputer(strategy = 'mean') #median , most_frequent ,constant
train_mean.fit(train)
train_imputed = train_mean.transform(train)
```

```python
train_imputed
```

### iterativeimputation

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
from sklearn.linear_model import LinearRegression
```

```python
imp = IterativeImputer(estimator = LinearRegression(), 
                       tol= 1e-10, 
                       max_iter=30, 
                       verbose=2, 
                       imputation_order='roman')
## estimater는 BayesianRidge방법이 default이고 툴은 바뀌가면서 돌려보면 될듯함슨
```

```python
pd.DataFrame(imp.fit_transform(train))##하다가 컴퓨터 터질꺼 같아서 멈췄삼
```
